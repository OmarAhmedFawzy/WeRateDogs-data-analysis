## Importing needed modules

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
```

## Reading data

```python
df = pd.read_csv('fav_count.csv')
df1 = pd.read_csv('num_dog.csv')
```

## Modifying data for model

```python
df['count'] = df1['count']
df.drop(columns={'Unnamed: 0'}, inplace=True)
df
```

## Encoding dog types

```python
label_encoder = LabelEncoder()
df['p1_encoded'] = label_encoder.fit_transform(df['p1'])
```

## Making target(y) and features(x) variable 

```python
y = df['favorite_count']
x = df[['p1_encoded','count']]
```

## Making model

```python
fav_count_model = DecisionTreeRegressor(random_state=1)
fav_count_model.fit(x, y)
```

## Testing model on data with knon answers

```python
fav_count_model.predict(x.head())
```
